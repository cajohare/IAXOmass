```python
from numpy import *
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import AxionFuncs

# init plot
plt.rcParams['axes.linewidth'] = 2.5
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
fig = plt.figure(figsize=(14.3,13))
ax = fig.add_subplot(111)

# Axion energy range for the plot
na = 10000
Ea = linspace(0.1,10,na)

# Rescale from 1/(keV m s) to 10^20/(keV cm year)
rs = (100**2.0)*(3600*24*365)/1e20

# Generate fluxes
dPhi_P = AxionFuncs.AxionFlux_Primakoff(1e-11,Ea)
dPhi_AR = AxionFuncs.AxionFlux_Axioelectron(1e-13,Ea)

# Plot
plt.plot(Ea,dPhi_P*rs,'-',linewidth=3,label=r'Axion-Photon $g_{a\gamma} = 10^{-11}$ GeV$^{-1}$')
plt.fill_between(Ea,dPhi_P*rs,color="Blue",linewidth=3,alpha=0.2)
plt.plot(Ea,dPhi_AR*rs,'-',linewidth=3,label=r'Axion-Electron $g_{ae} = 10^{-13}$')
plt.fill_between(Ea,dPhi_AR*rs,color="Orange",linewidth=3,alpha=0.2)

# Plot fiddling
ax.set_xlim(left=0.0, right=10.0)
ax.set_ylim(bottom=0.0)
ax.tick_params(which='major',direction='in',width=2,length=10,right=True,top=True)
ax.tick_params(which='minor',direction='in',width=1,length=7,right=True,top=True)
plt.xticks(fontsize=35)
plt.yticks(fontsize=35)
plt.xlabel(r"$E_a$ [keV]",fontsize=45);
plt.ylabel(r"$\frac{{\rm d}\Phi}{{\rm d}E_a}$ [$10^{20}$ keV$^{-1}$ cm$^{-2}$ yr$^{-1}$]",fontsize=45);
plt.legend(fontsize=35,frameon=False)

# Save figure
plt.show()
fig.savefig('../plots/SolarAxionFlux.pdf',bbox_inches='tight')
fig.savefig('../plots/plots_png/SolarAxionFlux.png',bbox_inches='tight')
```
